### Create a source & set up the detector

```python
### SET UP THE SOURCE ###
print("Set up the source")

# Source position and type
gvxr.setSourcePosition(-40.0,  0.0, 0.0, "cm")
gvxr.usePointSource()     # For a point source
# gvxr.useParallelBeam()  # For a parallel source

# Source spectrum
# gvxr.setMonoChromatic(0.08, "MeV", 1000)  # A monochromatic beam - 1000 photons of 80 keV (i.e. 0.08 MeV) per ray
kV = 120
max_number_of_energy_bins = 50

filtration = [["Cu", 1.0, "mm"]]

loadSpectrumXpecgen(kV, filters=filtration, th_in_deg=12, max_number_of_energy_bins=max_number_of_energy_bins);

energy_bins = gvxr.getEnergyBins("keV")
photon_count = np.array(gvxr.getPhotonCountEnergyBins(), dtype=np.single)
photon_count /= photon_count.sum()
```

```python
### SET UP THE DETECTOR ###
print("Set up the detector")

# Detector Position and direction
gvxr.setDetectorPosition(10.0, 0.0, 0.0, "cm")
gvxr.setDetectorUpVector(0, 0, -1)

# Detector Pixel definitions
gvxr.setDetectorNumberOfPixels(640, 320)
gvxr.setDetectorPixelSize(0.5, 0.5, "mm")
```

```python
gvxr.makeCylinder("Cylinder", 24, 4, 3, "cm")

gvxr.addPolygonMeshAsOuterSurface("Cylinder")
gvxr.setElement("Sphere", "Fe");
```

```python
x_ray_image = gvxr.computeXRayImage();

# Update the visualisation window
gvxr.displayScene();
```
